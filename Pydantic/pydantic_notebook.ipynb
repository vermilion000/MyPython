# **Pydantic :**
**Pydantic** is used for Dataparsing,DataValidation and Data Coercion.

BaseModel: Class provided in pydantic to be used as base class for our model.It has support for validation, serialization and parsing .

Model: Python class that imports baseclass from pydantic that will contain our fields and validations

Fields : variables of class that will validated with annotated type

Validation : Validation start as soon as instantiation is called for.

Note == fields can be accessed as normal instatce variables.

```python
#!pip install pydantic           #uncomment if pydantic is not installed
from typing import Optional
from pydantic import BaseModel,Field   #importing BaseModel Class

class Candidate(BaseModel):
  name:str        #this is called type annotation
  age :int = 20   #default value
  sex :Optional[int] = None#Optional field which can be optional argument
  weight : int = Field(ge = 45,le = 95)  #limiting field values for validation
  full_name:str = Field(pattern = r'^[a-zA-z]+$',min_length=1,max_length = 20) #String limits

c = Candidate(name = "Pablo", weight = 75,full_name = "PabloEscobar" )
print(c.name)
```

# **Custom Validation and validation-Error Handling:**
we can customize validators using @field_validator and @model_validator decorators for custom validatoion function
@field_validator : For single field
@model_validator :Multiple or Whole model validation

'try-except' for error handling

'raise' statment raises valiation error


```python
from pydantic import BaseModel,field_validator,model_validator   #importing BaseModel Class

class Candidate(BaseModel):
  name:str        #this is called type annotation
  age :int = 20   #default value
  address :str
  confirm_address :str
  @field_validator('age')
  def validate_age(cls,v):
    if v<15 :
      raise ValueError("Candidate Age must be greater than 15")
    return v
  @model_validator(mode = 'after')
  def address(cls,model):
    if model.address != model.confirm_address:
      if model.age > 25:
        raise ValueError("Address and Confirm Address must be same for Age above 25")
    return model
try :
  c = Candidate(name = "Pablo", address = '/Home',confirm_address = '/home')
except ValueError as e:
  print(e)
```

# **Serialization and Parsing :**

*   **model_validate()** for Parsing or

*   **model_dump()** for Serialisation of Python Dict/list

**JSON :**


*   **model_validate_json()** for Parsing or
deserialisation of JSON string
*   **model_dum_json()** for Serialisation of JSON Object

**Pickle :**


*   **pickle.load** for Parsing or
deserialisation Pickle String

# **Custom and Strict Datatype :**

Custom Fields : can be created for validation


```python
from pydantic import BaseModel,conint,constr,StrictBool,StrictInt,StrictStr

class Product(BaseModel):
    name: constr(min_length=1)
    price: conint(gt=0)
    brand :StrictStr()
p = Product(name = 'Oreo',price = 20,brand = 'Nabisco')
print(p)
```
